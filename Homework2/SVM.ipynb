```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sklearn
from keras._tf_keras.keras.datasets import fashion_mnist
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn import metrics
from sklearn.model_selection import GridSearchCV 
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
import sklearn.svm as svm
from sklearn.preprocessing import StandardScaler
from sklearn.inspection import DecisionBoundaryDisplay
```

```python
#Load data
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()
#Split the training data into training and validation
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

# Programming Assignment 1.1 Linear SVM for Fashion MNIST
train_result = []
test_result = []
validation_result = []
support_vectors_count = []

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train.reshape(X_train.shape[0], -1))
X_val_scaled = scaler.transform(X_val.reshape(X_val.shape[0], -1))
X_test_scaled = scaler.transform(X_test.reshape(X_test.shape[0], -1))

# As previously determined by the grid search, the best C value is 0.01, but we'll train the model with different C values
c_values = [0.0001, 0.001, 0.01, 0.1, 1, 10, 100, 1000, 10000]

for c in c_values:
    # Train the model
    print("Training model with C =", c)
    model = svm.LinearSVC(C=c, max_iter=1000)
    model.fit(X_train_scaled, y_train)
    print("Model trained with C =", c)
    # Predict and store results
    train_result.append(accuracy_score(y_train, model.predict(X_train_scaled)))
    validation_result.append(accuracy_score(y_val, model.predict(X_val_scaled)))
    print("Validation accuracy for C =", c, "is", validation_result[-1])
    test_result.append(accuracy_score(y_test, model.predict(X_test_scaled)))
    # Record number of support vectors
    decision_function = model.decision_function(X_train_scaled)
    support_vector_indices = np.where(np.abs(decision_function) <= 1 + 1e-15)[0]
    support_vectors = X_train_scaled[support_vector_indices]
    print("Number of support vectors for C =", c, "is", len(support_vectors))
    

# Plot the curves for training, validation, and test data for each C value
plt.title('Linear Kernel SVM: Accuracy vs C')
plt.plot(c_values, train_result, label='Training')
plt.plot(c_values, validation_result, label='Validation')
plt.plot(c_values, test_result, label='Test')
plt.xlabel('C parameter')
plt.ylabel('Accuracy')
plt.xscale('log')
plt.xticks(c_values)
plt.legend()
#plt.savefig('LinearSVM.jpeg', dpi=1200)
plt.show()

# Plot number of support vectors vs C values
plt.title('Linear Kernel SVM: Number of Support Vectors vs C')
plt.plot(c_values, support_vectors_count, marker='o')
plt.xlabel('C parameter')
plt.ylabel('Number of Support Vectors')
plt.xscale('log')
plt.xticks(c_values)
#plt.savefig('LinearSVM_SupportVectors.jpeg', dpi=1200)
plt.show()

```

```python
print("Best C value for validation data is", c_values[np.argmax(validation_result)])
print("Best C value for test data is", c_values[np.argmax(test_result)])
print("Highest accuracy for validation data is", max(validation_result))
```

```python
# Programming Assignment 1.2 Linear SVM for Fashion MNIST with best C value and confusion matrix
# Train the model with the best C value
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()

X_train_flat = X_train.reshape(X_train.shape[0], -1)
X_test_flat = X_test.reshape(X_test.shape[0], -1)

# Scale the data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_flat)
X_test_scaled = scaler.transform(X_test_flat)

# Model with the best C value
best_model = svm.LinearSVC(C=0.01, max_iter=1000, dual=False, tol=1e-3)

# Train the model
best_model.fit(X_train_scaled, y_train)

# Predict the test data
y_pred = best_model.predict(X_test_scaled)

# Score
accuracy = accuracy_score(y_test, y_pred)
print(accuracy)

# Build confusion matrix
confusion_mtx = confusion_matrix(y_test, y_pred)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix=confusion_mtx, display_labels=range(10))
cm_display.plot()
plt.title('Linear Kernel SVM: Confusion Matrix')
plt.savefig('LinearSVMConfusionMatrix.jpeg', dpi=1200)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

#Load data
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()
#Split the training data into training and validation
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

#Programming Assignment 1.3 Polynomial SVM for Fashion MNIST with best C value 
polynomial_degrees = [1, 2, 3, 4]
train_result = []
test_result = []
validation_result = []

scaler = StandardScaler()
X_train_scaled = X_train.reshape(X_train.shape[0], -1)
X_val_scaled = X_val.reshape(X_val.shape[0], -1)
X_test_scaled = X_test.reshape(X_test.shape[0], -1)

#Store the number of support vectors for each degree
support_vectors_count = []

C = 0.01
for degree in polynomial_degrees:
    # Train the model
    print("Training model with degree =", degree)
    if degree == 1:
        model = svm.LinearSVC(C=C, max_iter=1000)
    else:
        model = svm.SVC(kernel='poly', C=C, degree=degree, max_iter=1000)
    model.fit(X_train_scaled, y_train)
    print("Model trained with degree =", degree)
    if degree != 1:
        support_vectors = model.support_vectors_
        support_vectors_count.append(len(support_vectors))
    else: 
        decision_function = model.decision_function(X_train_scaled)
        support_vector_indices = np.where(np.abs(decision_function) <= 1 + 1e-15)[0]
        support_vectors = X_train_scaled[support_vector_indices]
        support_vectors_count.append(len(support_vectors))
    print("Number of support vectors for degree =", degree, "is", len(support_vectors))
    # Predict and store results
    train_result.append(accuracy_score(y_train, model.predict(X_train_scaled)))
    validation_result.append(accuracy_score(y_val, model.predict(X_val_scaled)))
    test_result.append(accuracy_score(y_test, model.predict(X_test_scaled)))
    print("Validation accuracy for degree =", degree, "is", validation_result[-1])

# Plot the curves for training, validation, and test data for each degree value
plt.title('Polynomial Kernel SVM: Accuracy vs Degree')
plt.plot(polynomial_degrees, train_result, label='Training')
plt.plot(polynomial_degrees, validation_result, label='Validation')
plt.plot(polynomial_degrees, test_result, label='Test')
plt.xlabel('Polynomial Degree')
plt.ylabel('Accuracy')
plt.legend()
plt.savefig('PolynomialSVM.jpeg', dpi=1200)
plt.show()


# Plot number of support vectors vs C values
plt.title('Polynomial Kernel SVM: Number of Support Vectors vs degree')
plt.plot(polynomial_degrees, support_vectors_count, marker='o')
plt.xlabel('C parameter')
plt.ylabel('Number of Support Vectors')
plt.xscale('log')
plt.xticks(polynomial_degrees)
plt.savefig('PolySVM_SupportVectors.jpeg', dpi=1200)
plt.show()
```
